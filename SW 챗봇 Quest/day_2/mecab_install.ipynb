### mecab 을 이용한 한국어 형태소 분석

```
from konlpy.tag import Mecab
import time
```

```
mecab=Mecab() 
text = '자연어는 처리가 필요합니다. 전처리가 중요하지요. 철수가 전처리를 합니다.'
```

```
start = time.time()
print(mecab.morphs(text))
print('\n')
print(mecab.pos(text))
print('\n')
print(mecab.nouns(text))
print(time.time() - start)
```
